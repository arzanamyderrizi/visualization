DSI Assignment 3 Text - Visualization 2 - School Board Expenditures and Revenues in Ontario

What software did you use to create your data visualization? R

Who is your intended audience?

Education policy analysts, school board administrators, researchers, and the general public interested in understanding how public education in Ontario is funded and how financial health has evolved over recent years. 

What information or message are you trying to convey with your visualization?

The visualization highlights the total revenue of Ontario’s District School Boards each year and overlays their surplus or deficit at year end, making it easy to see how well-funded school boards are and whether they consistently spend within their means.

What aspects of design did you consider when making your visualization? How did you apply them? With what elements of your plots?

I focused on:
- Simplicity: By showing only two key variables — revenue and surplus/deficit — side by side.
- Color contrast: Blue and pink were chosen for high visibility and aesthetic balance.
- Readability: Decimal points were removed from axis labels, and the plot avoids clutter by omitting text labels on bars.
- Gridline removal: This improves visual clarity and emphasizes the shapes and magnitudes of the bars.

How did you ensure that your data visualizations are reproducible? If the tool you used to make your data visualization is not reproducible, how will this impact your data visualization?

The entire visualization is created through a scripted workflow in R. It dynamically downloads the official Excel file from Ontario’s Open Data portal, reads and processes it, and builds the plot using ggplot2. This makes the visualization fully reproducible by anyone with access to R.

How did you ensure that your data visualization is accessible?

Accessibility was addressed through:
- Clear color choices that maintain contrast even for users with color vision deficiency.
- Easy to read font sizes and well-spaced bar elements.
- Minimal reliance on overlapping or small-text labels.
- Use of ggplot2's clear default theming, which works well with screen readers if exported to accessible formats.

Who are the individuals and communities who might be impacted by your visualization?

- Students and families, whose education is shaped by school board funding.
- Education policymakers, who rely on financial health metrics to allocate funding.
- Taxpayers and citizens, interested in accountability and transparency in school funding.
- School administrators and finance staff, who use surplus/deficit data to inform decisions.

How did you choose which features of your chosen dataset to include or exclude from your visualization?

I focused on two metrics that are widely applicable across all school boards, understandable without needing advanced financial knowledge, and key indicators of fiscal health. I excluded expenses and accumulated surplus to avoid clutter and keep the focus on year-to-year fiscal performance.

What ‘underwater labour’ contributed to your final data visualization product?

- Identifying the correct data file and variable names from a large Excel workbook.
- Understanding the structure and meaning of multiple financial indicators.
- Cleaning, reshaping, and testing different plotting styles.
- Troubleshooting axis formatting, bar spacing, and color balance to find the most effective and clear presentation.
