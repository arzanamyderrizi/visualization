DSI Assignment 3 Text - Visualization 1 - School Enrollment in Ontario

What software did you use to create your data visualization? Python

Who is your intended audience? 

Education policy analysts, school board administrators, researchers, and the general public interested in understanding the distribution of student enrolment across Ontario’s education systems at the elementary and secondary levels.

What information or message are you trying to convey with your visualization?

The visualization shows the relative share of enrolment between Public and Roman Catholic school
systems for both elementary and secondary levels in Ontario during the 2023–24 academic year. It emphasizes the dominance of the Public system while allowing users to explore the exact percentage shares interactively.

What aspects of design did you consider when making your visualization? How did you apply them? With what elements of your plots?

I considered:
- Color contrast: I used distinct, colorblind-accessible colors to clearly differentiate between systems.
- Clean layout: Gridlines were removed, margins adjusted, and labels made concise to reduce visual clutter.
- Interaction: Tooltips were customized to show only percentages on hover, keeping the visual minimal while preserving access to key information.
- Axis formatting: The y-axis uses comma-separated values to improve readability for large numbers.

How did you ensure that your data visualizations are reproducible? If the tool you used to make your data visualization is not reproducible, how will this impact your data visualization?

The entire workflow is coded in Python and uses open government data from Ontario's publicly 
available ZIP files. The code dynamically:
- Downloads and extracts the Excel file,
- Cleans and filters the data,
- Computes enrolment shares, and
- Plots the figure.

Because everything is scripted and version controlled, anyone can reproduce the visualization by running the code in a Jupyter notebook. This enhances transparency and trust in the findings.

How did you ensure that your data visualization is accessible?

I used high-contrast, distinguishable colors for users with visual impairments; Large, fonts and avoided excessive labels; Interactive elements instead of fixed text to reduce clutter while remaining keyboard- and screen-reader-friendly through Plotly’s output.

Who are the individuals and communities who might be impacted by your visualization?

- Students and families navigating public vs. Catholic school options,
- School boards assessing system enrolment pressures,
- Policy makers involved in funding allocation or reviewing education delivery models,
- Advocacy groups promoting equity across school systems.

How did you choose which features of your chosen dataset to include or exclude from your visualization?

I focused only on total enrolment by system and level (Public/Roman Catholic × Elementary/
Secondary) to simplify the message and target the most policy-relevant breakdowns.

What ‘underwater labour’ contributed to your final data visualization product?

Significant unseen work included:
- Finding and downloading the correct ZIP file from Ontario’s Open Data portal,
- Exploring and understanding the structure of the Excel files,
- Cleaning and reshaping the data to isolate just the needed categories,
- Normalizing system names, and
- Testing different formats, axes, and hover options to find a clean and effective visual presentation.
